# Analiza poszczególnych zmiennych.

```python
df.hist() #wizualizacja za pomocą histogramu
```

```python
df = df.rename(columns={'długość kielicha (sepal length)': 'kielich_długość', 'szerokość kielicha (sepal width)' : 'kielich_szerokość', 'długość płatka (petal length)' : 'płatek_długość', 'szerokość płatka (petal width)' : 'płatek_szerokość', 'klasa (class)' : 'klasa'})
df # zmiana nazw kolumn
```

```python
df['pole_kielicha']= df['kielich_długość'] * df['kielich_szerokość'] #obliczmy pole kielicha
df
```

```python
df['pole_płatka']= df['płatek_długość'] * df['płatek_szerokość'] #obliczmy pole płatka dla wszystkich rekordów
df
```

```python
statystyki_kielich=df.groupby('klasa', as_index=False).agg({
    'kielich_długość':['mean', 'median',  'min', 'max'],
    'kielich_szerokość':['mean', 'median',  'min', 'max'],
    'pole_kielicha' : ['mean', 'median',  'min', 'max']
}).copy() 
# obliczamy wartość średniej, mediana, miminum, maximum dla długości i szerokości kielicha dla poszczególnych klas irisa

statystyki_kielich
```

```python
statystyki_kielich.plot(kind='bar', x='klasa', y=['kielich_długość','kielich_szerokość', 'pole_kielicha',])
```

```python
statystyki_płatek=df.groupby('klasa', as_index=False).agg({
    'płatek_długość':['mean', 'median',  'min', 'max'],
    'płatek_szerokość':['mean', 'median',  'min', 'max'],
    'pole_płatka' : ['mean', 'median',  'min', 'max']
}).copy() 
# obliczamy wartość średniej, mediana, miminum, maximum dla długości i szerokości płatka dla poszczególnych klas irisa

statystyki_płatek
```

```python
statystyki_płatek.plot(kind='bar', x='klasa')
```
